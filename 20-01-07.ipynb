```python
from pylab import plt
plt.style.use('ggplot')
import matplotlib as mpl
mpl.rcParams['font.family']='바탕'
def f(x):
    return np.sin(x) + 0.5 *x     #주기함수
x =np.linspace(-2*np.pi, 2*np.pi,50)
plt.plot(x,f(x), 'b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
#최소제곱법을 이용해서 fitting 적합
```

```python
#polyfit 다차방정식으로 fitting(적합): 계수를 찾아내는 것
reg = np.polyfit(x,f(x), deg=5) #5차 방정식으로 피팅하라.deg에 숫자가 2를 넣으면 2차 방정식이 나옴
ry = np.polyval(reg,x)          #5차방정식으로 회귀

plt.plot(x,f(x), 'b',label='f(x)') #원본데이터
plt.plot(x, ry,'r.', label='regression') #최소제곱법으로 피팅된 데이터
plt.legend(loc=0)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
import pandas as pd
data={'지역' : ['수도권','경상권','수도권','경상권'],
     '2018' :[9904312,3448737,2890451,2466052],
     '2017' :[9631482,3393191,2632035,2431774],
     '2016' :[9762546,3512547,2517680,2456016],
     '2015' :[9853972,3655437,2466338,2473990],
     '2015-2018증가율' :[0.0283, 0.0163, 0.0982, 0.0141]}
frame1 = pd.DataFrame(data,columns=['지역','2018','2017','2016','2015','2015-2018증가율'],index=['서울','부산','인천','대구'])
print(frame1) 
```

```python
frame1['2015-2018증가율']=frame1['2015-2018증가율']*100
frame1
```
